# 3. DATA TRANFORMATION 

```python

df_option = pd.read_json('option_data.json',orient='records', lines=True)

# Atribute a unique_id to facilite the merge 
df_option["unique_id"] = df_option.index
dict_option = df_option.to_dict("records")


## Tree column manipulation
df_tree = json_normalize(dict_option,'tree',['unique_id'], sep='_').sort_values(['unique_id', 'depth'], ascending=[True, True])

df_option_path = df_tree.groupby(['unique_id'], as_index = False).agg({'id': ';'.join})
df_option_path = df_option_path.rename(columns={"id": "option_path"})


## Metadata column manipulation
df_metadata = json_normalize(dict_option,'metadata',['unique_id'], sep='_')

sap_filter = df_metadata['key']=="sap"
tmid_c_filter = df_metadata['key']=="tmid_c"

df_metadata_sap = df_metadata[sap_filter][['value', 'unique_id']].rename(columns={'value':'sap'})
df_metadata_tmid_c = df_metadata[tmid_t_filter][['value', 'unique_id']].rename(columns={'value':'tmid_c'})


## Merge dataframes

df_default_columns = df_option[['option_id', 'option_value', 'unique_id']]  

df_merged = pd.merge(df_default_columns, df_option_path, how='left', on='unique_id')
df_merged = pd.merge(df_merged, df_metadata_sap, how='left', on='unique_id')
df_merged = pd.merge(df_merged, df_metadata_tmid_c, how='left', on='unique_id')

df_merged = df_merged.drop(columns=['unique_id']).reset_index()

df_merged.head()

```
